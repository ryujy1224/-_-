```python
graph = {
    # 예시 그래프: 친구 관계
    '류재윤': ['이희원', '박정우'],
    '이희원': ['친구1', '친구2'],
    '박정우': ['친구3'],
    '친구1': [],
    '친구2': ['친구3'],
    '친구3': []
}

from collections import deque

def dfs(graph, v, visited):
    visited.add(v)
    print(v, end=' ')
    for neighbor in graph[v]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    while queue:
        v = queue.popleft()
        print(v, end=' ')
        for neighbor in graph[v]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    print()

def recommend_friend(graph, start):
    visited = set()
    queue = deque([(start, 0)])

    print(f"{start}의 친구 추천 (2단계):")

    while queue:
        current, depth = queue.popleft()

        if current not in visited:
            visited.add(current)

            if depth == 2:
                print("-", current)

            if depth < 2:
                for neighbor in graph[current]:
                    queue.append((neighbor, depth + 1))
    print()

START_NODE = '류재윤'

print("DFS 탐색 순서:")
dfs(graph, START_NODE, set())
print()

print("BFS 탐색 순서:")
bfs(graph, START_NODE)

recommend_friend(graph, START_NODE)
```
