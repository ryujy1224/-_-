```python
import networkx as nx
import matplotlib.pyplot as plt

# 1.    (  )
# Ryu Jaeyun, Park Jungwoo, Lee Heewon, Friend1, Friend2, Friend3
graph = {
    'Ryu Jaeyun': ['Park Jungwoo', 'Lee Heewon', 'Friend1', 'Friend2'],
    'Park Jungwoo': ['Ryu Jaeyun', 'Friend3'],
    'Lee Heewon': ['Ryu Jaeyun'],
    'Friend1': ['Ryu Jaeyun', 'Friend3'],
    'Friend2': ['Ryu Jaeyun'],
    'Friend3': ['Park Jungwoo', 'Friend1']
}

# 2.   
G = nx.Graph()
for person, friends in graph.items():
    for friend in friends:
        G.add_edge(person, friend)

# 3.  
pos = nx.spring_layout(G, k=0.5, iterations=50, seed=15)

# 4.  
plt.figure(figsize=(10, 7))

nx.draw(G, pos,
        with_labels=True,
        node_color='lightgreen',
        node_size=2500,
        font_size=12,
        font_color='black',
        font_weight='bold',
        edge_color='gray',
        linewidths=1,
        width=1.5
)

plt.title("Friendship Network (Undirected)", fontsize=16)
plt.show()
```
